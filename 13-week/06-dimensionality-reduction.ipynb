```python
from sklearn import datasets
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
X, y = datasets.make_classification(
    n_samples=300,
    n_features=3,
    n_classes=3,
    n_redundant=0,
    n_clusters_per_class=1,
    weights=[0.5, 0.3, 0.2],
    random_state=42,
)
pca = PCA(n_components=2, svd_solver="randomized")
X_fitted = pca.fit(X).transform(X)
fit = pca.fit(X)

print(f"Explained Variance: {fit.explained_variance_ratio_}")
```

```python
# Plot original data
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1, projection="3d")
ax.scatter(xs=X[:, 0], ys=X[:, 1], zs=X[:, 2], c=y)
ax.set_title("Original 3-feature data")
ax.set_label("X0")
ax.set_label("X1")
ax.set_label("X2")
plt.show()

# Plot reduced data
fig, ax = plt.subplots(figsize=(9, 6))
plt.title("Reduced 2-feature data")
plt.xlabel("X_fitted_0", fontsize=20)
plt.ylabel("X_fitted_1", fontsize=20)
plt.scatter(X_fitted[:,0], X_fitted[:,1], s=50, c=y)
```
