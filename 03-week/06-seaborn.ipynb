### Scatter Plots

- A scatter plot is a type of chart that shows the relationship between two different variables.
- Think of a scatter plot as a way to visually see how two things are related.
- For example, you can plot people's height on one axis and their weight on the other. Each point on the scatter plot represents one person, showing their height and weight together.
- A scatter plot is useful when:
  - We want to see relationship b/w 2 variables.
  - We want to detect patterns or trends (if one variable increases then other increases +ve trend or if one decreases when other increases -ve trend).

```python
# Syntax for Scatter Plot

# 1. Create grid in the plot
sns.set_theme(style="darkgrid")

# 2. Load the data
attrition = pd.read_csv("./csv-data/Attrition.csv")
# print(attrition.columns)
# print(attrition.dtypes)
# print(attrition.shape[0])

# attrition.info()
# attrition.describe()
# attrition.head()

attrition.head()
```

```python
sns.scatterplot(x="Age", y="MonthlyIncome", data=attrition)
plt.show()
```

### What is KDE?

- KDE stands for Kernel Density Estimate.
- It is a method for estimating the probability density function (PDF) of a continuous random variable.
- It is a way to smooth out the distribution of your data to make it easier to visualize.
- Unlike the histogram, which divides data into discrete bins, KDE provides a smooth curve that represents the density of the data.
- The kernel is a function (typically Gaussian) that spreads each data point into a smooth curve.
- The bandwidth is a parameter that controls the smoothness of the KDE. A smaller bandwidth will lead to a curve that fits the data closely (but might be noisy), while a larger bandwidth will produce a smoother curve (but might miss some details of the distribution).

```python
# Generate random data
data = np.random.randn(1000)

# Create a plot with different bandwidths
plt.figure(figsize=(10, 6))

# Plot with small bandwidth (detailed, less smooth)
sns.kdeplot(data, bw_adjust=0.5, label="Small Bandwidth (0.5)", color="blue")

# Plot with default bandwidth
sns.kdeplot(data, label="Default Bandwidth", color="green")

# Plot with large bandwidth (smoother, less detailed)
sns.kdeplot(data, bw_adjust=2, label="Large Bandwidth (2)", color="red")

# Add title and legend
plt.title("Effect of Bandwidth on KDE")
plt.legend()
plt.show()
```

```python
# Generate random data from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)

# Plot the Gaussian distribution
sns.set_theme(style="whitegrid")
sns.histplot(data, kde=True, color="skyblue")

# Add title and labels
plt.title("Gaussian Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```
