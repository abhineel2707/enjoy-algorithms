### What is Seaborn?
- Seaborn is a data visualization library built on top of matplotlib and closely integrated with pandas data structures in Python.
- While Matplotlib is a powerful and flexible library for plotting in Python, Seaborn adds several benefits that make it particularly useful.
- Seaborn offers a simplified syntax for creating common plots such as bar charts, line plots, scatter plots, histograms, and more.
- It abstracts many of the details required in Matplotlib, making it easier and faster to create complex visualizations with just a few lines of code.
- Seaborn's default styles and color palettes are more visually appealing and consistent than Matplotlib's default settings.
- Seaborn is designed to work seamlessly with Pandas DataFrames, allowing you to easily plot data directly from DataFrame columns without needing to extract or reshape the data manually.

```python
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
```

### Line Plot

```python
data_1 = [1,2,3,4,5,6,7]
data_2 = [2,3,4,1,5,6,7]
```

```python
plt.plot(data_1, data_2)
plt.show()
```

```python
dataframe_1 = pd.DataFrame(
    {
        "data_1":data_1,
        "data_2":data_2
    }
)
sns.lineplot(x="data_1", y="data_2",data=dataframe_1)
plt.show()
```

```python
data_1 = [1,2,3,4,5,6,7]
data_2 = [2,3,4,1,5,6,7]

sns.lineplot(x=data_1, y=data_2)
plt.show()
```

```python
penguins_data = sns.load_dataset("penguins")
penguins_data
```

```python
sns.lineplot(x="bill_depth_mm", y="flipper_length_mm", data=penguins_data)
plt.show()
```

```python
sns.lineplot(x="bill_length_mm", y="bill_depth_mm", data=penguins_data, hue="sex")
plt.show()
```

```python
sns.lineplot(x="bill_length_mm", y="bill_depth_mm", data=penguins_data, hue="sex", size=10)
plt.show()
```

```python
sns.lineplot(x="bill_length_mm", y="bill_depth_mm", data=penguins_data, hue="sex", style="sex")
plt.show()
```

```python
sns.lineplot(x="bill_length_mm", y="bill_depth_mm", data=penguins_data, hue="sex", style="sex", palette="Accent")
plt.show()
```

```python
sns.lineplot(x="bill_length_mm", y="bill_depth_mm", data=penguins_data, hue="sex", style="sex", palette="Accent", markers=["o",">"])
plt.show()
```

```python
penguins_less_data = sns.load_dataset("penguins").head(20)
penguins_less_data
```

```python
sns.lineplot(x="bill_length_mm", y="bill_depth_mm", data=penguins_less_data, hue="sex", style="sex", palette="Accent", markers=["o",">"])
plt.show()
```

```python
sns.lineplot(x="bill_length_mm", y="bill_depth_mm", data=penguins_less_data, hue="sex", style="sex", palette="Accent", markers=["o",">"], dashes=False)
plt.show()
```

```python
sns.lineplot(x="bill_length_mm", y="bill_depth_mm", data=penguins_less_data, hue="sex", style="sex", palette="Accent", markers=["o",">"],legend="full",dashes=False)
plt.show()
```

```python
sns.lineplot(x="bill_length_mm", y="bill_depth_mm", data=penguins_less_data, hue="sex", style="sex", palette="Accent", markers=["o",">"],legend=False,dashes=False)

plt.show()
```

```python
sns.lineplot(x="bill_length_mm", y="bill_depth_mm", data=penguins_less_data, hue="sex", style="sex", palette="Accent", markers=["o",">"],legend=False,dashes=False)
plt.grid()
plt.title("Penguin Data")
plt.show()
```
