```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
mpl.style.available
```

```python
mpl.style.use('seaborn-v0_8-darkgrid')
```

```python
fig, ax = plt.subplots()
```

```python
x_pts = np.linspace(-2*np.pi, 2*np.pi, 200)
y_pts = np.sin(x_pts)
ax.plot(x_pts, y_pts, label='sin')
fig
```

```python
x_pts = np.linspace(-2*np.pi, 2*np.pi, 200)
fig, ax = plt.subplots()
ax.plot(x_pts, np.sin(x_pts), label='sin')
ax.plot(x_pts, np.cos(x_pts), label='cos')
ax.set_xlabel('x-axis label')
ax.set_ylabel('y-axis label')
ax.set_title('Trig Functions')
ax.legend()
None
```

```python
x_pts = np.linspace(-2*np.pi, 2*np.pi, 200)
fig, ax = plt.subplots(figsize=(12,6))
ax.plot(x_pts, np.sin(x_pts), label='sin')
ax.plot(x_pts, np.cos(x_pts), label='cos')
ax.set_xlabel('x-axis label')
ax.set_ylabel('y-axis label')
ax.set_title('Trig Functions')
ax.legend()
None
```

```python
# Square of numbers
x_nums = np.arange(1,9,1)
y_nums = [num**2 for num in x_nums]
# print(x_nums, y_nums)

fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(x_nums, y_nums, label='square')
ax.set_xlabel('num')
ax.set_ylabel('square(num)')
ax.set_title('Square Function')
ax.legend()
None
```

```python
x_pts = np.linspace(-2*np.pi, 2*np.pi, 200)
nrows = 1
ncols = 2
fig, (ax1, ax2) = plt.subplots(nrows, ncols,figsize=(12,6))
ax1.plot(x_pts, np.sin(x_pts), label='sin')
ax2.plot(x_pts, np.cos(x_pts), label='cos')
ax1.set_xlabel('x-axis')
ax1.set_ylabel('sin')
ax2.set_xlabel('x-axis')
ax2.set_ylabel('cos')
ax1.set_title('sin')
ax2.set_title('cos')
ax1.legend()
ax2.legend()
```
