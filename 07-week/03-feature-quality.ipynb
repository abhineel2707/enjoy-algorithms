```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
purchase = pd.read_csv(
    "./csv-files/purchase.csv",
    header=0,
    usecols=["Price", "Payment Method", "Quantity"],
)

purchase
```

```python
# Replace missing values with mean
purchase = pd.read_csv(
    "./csv-files/purchase.csv",
    header=0,
    usecols=["Price", "Payment Method", "Quantity"],
)

purchase["Price"] = purchase["Price"].fillna(purchase["Price"].mean())
purchase
```

```python
# Replace missing values with median

purchase = pd.read_csv(
    "./csv-files/purchase.csv",
    header=0,
    usecols=["Price", "Payment Method", "Quantity"],
)

purchase["Price"] = purchase["Price"].fillna(purchase["Price"].median())
purchase
```

```python
# Replace missing value with backward filling

purchase = pd.read_csv(
    "./csv-files/purchase.csv",
    header=0,
    usecols=["Price", "Payment Method", "Quantity"],
)

purchase["Price"] = purchase["Price"].bfill()
purchase
```

```python
# Replace missing value with forward filling
purchase = pd.read_csv(
    "./csv-files/purchase.csv",
    header=0,
    usecols=["Price", "Payment Method", "Quantity"],
)

purchase["Price"] = purchase["Price"].ffill()
purchase
```

```python
# Replacing missing values with interpolation method
purchase = pd.read_csv(
    "./csv-files/purchase.csv",
    header=0,
    usecols=["Price", "Payment Method", "Quantity"],
)

purchase["Price"] = purchase["Price"].interpolate()
purchase
```

```python
# Replacing missing values with mode
purchase = pd.read_csv(
    "./csv-files/purchase.csv",
    header=0,
    usecols=["Price", "Payment Method", "Quantity"],
)

purchase["Payment Method"] = purchase["Payment Method"].fillna(
    purchase["Payment Method"].mode().iloc[0]
)
purchase
```

```python
# Replacing missing values with a new class of "Missing"
purchase = pd.read_csv(
    "./csv-files/purchase.csv",
    header=0,
    usecols=["Price", "Payment Method", "Quantity"],
)

purchase["Payment Method"] = purchase["Payment Method"].fillna("Missing")
purchase
```
